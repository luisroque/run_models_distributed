```python
import gpforecaster as gpf
import tsaugmentation as tsag
import matplotlib.pyplot as plt
import numpy as np
```

```python
groups = tsag.preprocessing.PreprocessDatasets('prison').apply_preprocess()
n = groups['predict']['n']
s = groups['train']['s']
```

```python
gpf_model = gpf.model.GPF('prison', groups)
model, like = gpf_model.train()
mean, lower, upper = gpf_model.predict(model, like)
```

```python
plt.plot(groups['predict']['data'].reshape((32, 48))[0]);
plt.plot(mean[0, :, 0])
plt.fill_between(np.arange(48), 
                 lower[0, :, 0], 
                 upper[0, :, 0], 
                 alpha=0.3);
```
